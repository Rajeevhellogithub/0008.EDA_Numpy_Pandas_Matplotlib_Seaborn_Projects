## Building plots dashboards
- Seaborn Kdeplot and Violinplot in Matplotlib Subplots

```python
# Building plots dashboards

sns.set_style('darkgrid')
f, axes = plt.subplots(2, 2, figsize=(15,10) )

k1 = sns.kdeplot(x=movies.BudgetMillions, y=movies.AudienceRating, ax=axes[0,0] )
k2 = sns.kdeplot(x=movies.BudgetMillions, y=movies.CriticRating, ax=axes[0,1] )

#k1.set( xlim=(-30, 300) )
#k2.set( xlim=(-30, 300) )

k3a = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, shade=True, shade_lowest=False, cmap='Reds', ax=axes[1,0] )
k3b = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, cmap='Reds', ax=axes[1,0] )

z = sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year', y='CriticRating', ax=axes[1,1] )

plt.show()
```

# Seaborn Jointplot

```python
# There are 559 datapoints in below plot
# Each datapoint is a pair of CriticRating-AudienceRating
# Basically jointplot is a scatterplot & it finds the relations b/w audience's & critic's ratings
# Also you can find the 'uniform distribution' (critics rating) and 'normal distriution' (audiences rating)
# Audiences rating is more dominant then critics rating
# Based on this we find out as most people are most liklihood to watch audience rating & less likely to watch critics rating
# let me explain the excel - if you filter audiences ratings & critics ratings. critics ratings has very low sum value compare to audiences ratings 

j1 = sns.jointplot(data=movies, x='CriticRating', y='AudienceRating')
```

# Seaborn Kdeplot
- Kernel Density Estimate Plot
- A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.
- KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
# find area of more density
# observe how desnity is distributed in plot
# kernel is center point and called as KDE
# dots are visualising in form of mingled lines
# observe the spreads/distributions of 'AudienceRating' and 'CriticRating'

sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating)
```
