# Boxplot or WhiskerPlot

**Box plot was was first introduced in year 1969 by Mathematician John Tukey.  
Box plot give a statical summary of the features being plotted.  
Top line represent the max value.  
Top edge of box is third Quartile.  
Middle edge represents the median.  
Bottom edge represents the first quartile value.  
The bottom most line respresent the minimum value of the feature.  
The height of the box is called as Interquartile range.  
The black dots on the plot represent the outlier values in the data.**

```python
fig = plt.gcf()
fig.set_size_inches(7,4)

fig = sns.boxplot(data=iris, x='Species', y='PetalLengthCm', hue='Species',
                  order=['Iris-virginica','Iris-versicolor','Iris-setosa'],
                  linewidth=0.5, orient='v', dodge=False
                 )
```

# Violin Plot

 - It is used to visualize the distribution of data and its probability distribution.
 - This chart is a combination of a Box Plot and a Density Plot.
 - Both combination is rotated and placed on each side to show the distribution shape of the data.
 - The thick black bar in the centre represents the interquartile range,
 - The thin black line extended from it represents the 95% confidence intervals,
 - The white dot is the median.
 - Box Plots are limited in their display of the data, as their visual simplicity tends to hide significant details about how values in the data are distributed

```python
fig=plt.gcf()
fig.set_size_inches(8, 4)
fig=sns.violinplot(data=iris, x='Species', y='SepalLengthCm', hue='Species')
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
sns.violinplot(data=iris, x='Species', y='PetalLengthCm', hue='Species')

plt.subplot(2,2,2)
sns.violinplot(data=iris, x='Species', y='PetalWidthCm', hue='Species')

plt.subplot(2,2,3)
sns.violinplot(data=iris, x='Species', y='SepalLengthCm', hue='Species')

plt.subplot(2,2,4)
sns.violinplot(data=iris, x='Species', y='SepalWidthCm', hue='Species')
```

# Heat Map

# Stacked Histogram

 - With Stacked Histogram we can see the distribution of SepalLengthCm of Different Species together.
 - This shows us the range of SepalLengthCm for the three different Species of Iris Flower.

```python
iris['Species'] = iris['Species'].astype('category')
iris.dtypes
```

```python
iris.Species.cat.categories
```

```python
sepal_len = list()
iris_species = list()

for i in iris.Species.cat.categories:
    sepal_len.append(iris[iris.Species==i].SepalLengthCm)
    iris_species.append(i)
```

```python
print(iris_species)
```

```python
fig = plt.hist(sepal_len, bins=30, stacked=True, rwidth=1, label=iris_species)

plt.legend()
plt.show()
```
